```python
# 2. 라이브러리 불러오기
import pandas as pd
import pingouin as pg
```

```python
# 3. 데이터 업로드
from google.colab import files
uploaded = files.upload()
```

```python
# 파일 불러오기 (업로드한 파일명에 맞춰 수정)
df = pd.read_csv("combined_sense_applicability_data.csv")
```

```python
# 4. 혼합분산분석 실행
anova = pg.mixed_anova(dv='applicability_score',
                       within='sense_type',
                       between='model_type',
                       subject='participant_id',
                       data=df)
print("### Mixed-design ANOVA 결과 ###")
print(anova)
```

```python
# 5. 사후 검정 (Bonferroni 보정 포함)
posthoc_models = pg.pairwise_ttests(dv='applicability_score',
                                    between='model_type',
                                    within='sense_type',
                                    subject='participant_id',
                                    data=df,
                                    padjust='bonf')
print("\n### 사후 검정 결과 ###")
print(posthoc_models)
```

```python
# 6. 결과 파일로 저장
anova.to_csv("mixed_anova_results.csv")
posthoc_models.to_csv("posthoc_results.csv")

print("결과가 mixed_anova_results.csv와 posthoc_results.csv 파일로 저장되었습니다.")
```
